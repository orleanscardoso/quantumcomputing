```python
import math

def BS_eta(alfa, beta, eta):
    alfa_out = alfa * math.sqrt(eta) - beta * math.sqrt(1 - eta)
    beta_out = alfa * math.sqrt(1 - eta) + beta * math.sqrt(eta)
    return alfa_out, beta_out
```

```python

import itertools
import numpy as np
import sys
def combinacoes(alfa, x):
# Gera todas as combinações possíveis de 0 e 1 para 3 entradas
    combinacoes = list(itertools.product([-alfa, alfa], repeat=6))
# Cria a matriz com as combinações
    M_in = [list(comb) for comb in combinacoes]
# Imprime a matriz
    M_in = np.array(M_in)
    return M_in
```

```python
import sympy
import itertools
import numpy as np
import sys
from IPython.display import display, HTML
import sys
# circuito1
alfa = sympy.symbols('alfa')
M_in = combinacoes(alfa, 6)

aux1 = M_in[:, 0]
aux2 = M_in[:, 1]
aux3 = M_in[:, 2]
aux4 = M_in[:, 3]
aux5 = M_in[:, 4]
aux6 = M_in[:, 5]


#(2)
aux2,aux1 = BS_eta(aux2, aux1, 1/2);
aux4,aux3 = BS_eta(aux4, aux3, 1/2);
aux6,aux5 = BS_eta(aux6, aux5, 1/2);
aux4 = -aux4;
aux5 = -aux5;
#(4)
aux1,aux3 = BS_eta(aux1, aux3, 1/2);
aux5,aux2 = BS_eta(aux5, aux2, 1/2);
aux4,aux6 = BS_eta(aux4, aux6, 1/2);
aux6 = -aux6;
aux2 = -aux2;

aux1 = aux1.reshape(-1, 1)
aux2 = aux2.reshape(-1, 1)
aux3 = aux3.reshape(-1, 1)
aux4 = aux4.reshape(-1, 1)
aux5 = aux5.reshape(-1, 1)
aux6 = aux6.reshape(-1, 1)

M_out = np.column_stack((aux1, aux2, aux3, aux4, aux5, aux6))

html_table = "<table border='1'>"

# Adicionar linhas à tabela
for linha in M_out :
    html_table += "<tr>"
    for elemento in linha:
        html_table += "<td>{}</td>".format(elemento)
    html_table += "</tr>"

# Fechar a tabela HTML
html_table += "</table>"

# Exibir a tabela HTML no Jupyter Notebook
display(HTML(html_table))
```
