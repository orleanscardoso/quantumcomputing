```python
def mapa(S_binario ):
    alfa = sympy.symbols('alfa')
    A4 = [0, 0, 1, 1]
    B4 = [0, 1, 0, 1]
    A8 = [0, 0, 0, 0, 1, 1, 1, 1]
    B8 = [0, 0, 1, 1, 0, 0, 1, 1]
    C8 = [0, 1, 0, 1, 0, 1, 0, 1]

    # Mapeamento de -1.0*alfa para 0 e 1.0*alfa para 1
    #S_binario = [0 if x == -1.0*alfa else 1 for x in S]

    # Se a quantidade de elementos for diferente de 4 ou 8, retornar None
    if len(S_binario) not in [4, 8]:
        return None

    # Inicializa uma lista para armazenar os termos da expressão
    expressao = []

    if len(S_binario) == 8:
        # Verifica quais variáveis são 1 em S e adiciona os termos correspondentes à lista de expressões
        if S_binario == A8:
            return 'A'
        elif S_binario == [1 - x for x in A8]:
            return "A'"
        elif S_binario == B8:
            return 'B'
        elif S_binario == [1 - x for x in B8]:
            return "B'"
        elif S_binario == C8:
            return "C"
        elif S_binario == [1 - x for x in C8]:
            return "C'"
        else:
            if S_binario[0] == 1:
                expressao.append("A'B'C' ")
            if S_binario[1] == 1:
                expressao.append("A'B'C ")
            if S_binario[2] == 1:
                expressao.append("A'BC'")
            if S_binario[3] == 1:
                expressao.append("A'BC")
            if S_binario[4] == 1:
                expressao.append("AB'C'")
            if S_binario[5] == 1:
                expressao.append("AB'C")
            if S_binario[6] == 1:
                expressao.append("ABC'")
            if S_binario[7] == 1:
                expressao.append("ABC")
        if expressao == "A'B'C'  + AB'C' + AB'C + ABC' + ":
            return 1
        if not expressao: 
            return 0
        return " + ".join(expressao)
    else:
        # Verifica quais variáveis são 1 em S e adiciona os termos correspondentes à lista de expressões
        if S_binario == A4:
            return 'A'
        elif S_binario == [1 - x for x in A4]:
            return "A'"
        elif S_binario == B4:
            return 'B'
        elif S_binario == [1 - x for x in B4]:
            return "B'"
        else:
            if S_binario[0] == 1:
                expressao.append("A'B'")
            if S_binario[1] == 1:
                expressao.append("A'B")
            if S_binario[2] == 1:
                expressao.append("AB'")
            if S_binario[3] == 1:
                expressao.append("AB")
            # Se a expressão for A'B' + A'B + AB' + AB, retorne 1
        if expressao == ["A'B'", "A'B", "AB'", "AB"]:
            return '1'
        # Se a expressão estiver vazia, retorne 0
        if not expressao:
            return '0'    
        return " + ".join(expressao)
```
